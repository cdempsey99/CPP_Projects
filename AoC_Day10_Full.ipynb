```python
import numpy as np
import statistics as st

#Read in the example and real inputs:
with open('AoC10example.txt') as input:
    lines = input.readlines()
with open('AoC10.txt') as input1:
    real_lines = input1.readlines()

#Some config variables:
#(i) The close brackets that correspond to the opens
corr = {
    "(" : ")",
    "[" : "]",
    "{" : "}",
    "<" : ">"
}

#(ii) The points value of each bracket type for Part 1
scores1 = {
    ')' : 3,
    ']' : 57,
    '}' : 1197,
    '>' : 25137
}

#(iii) The points value for Part 2
scores2 = {
    ')' : 1,
    ']' : 2,
    '}' : 3,
    '>' : 4
}

#(iv) List of open bracket characters
opens = ['(', '[', '{', '<']

#Function which loops over the list and removes adjacent pairs of open and close brackets (Part 1)
#Takes two inputs: (i) the list of characters which makes up the line and (ii) a list of open brackets
#(It only does loops once, so nested pairs will not be removed)
def remove_adjacents(x, opns):
    for i in range(len(x) - 1):
        #check for adjacent pairs and replace them with underscores:
        if x[i] in opns:
            if x[i+1] == corr[x[i]]:
                x[i] = '_'
                x[i+1] = '_'
    #now remove underscores:
    x = list(filter(('_').__ne__, x))
    return x

#Function to return (i) a list of brackets which consitute corruptions (for Part 1) and 
#(ii) the input list with all the open and close bracket pairs removed (for Part 2)
#This function again takes two inputs:A list (the line of characters) and a list of open brackets (Part 1)
def corrupted(x, opns):
    #convert string input to a list and get rid of trailing \n chars:
    x = list(x)
    for item in x:
        if item == '\n':
            x.remove(item)
    
    #remove all adjacent pairs repeatedly:
    len_i = len(x)
    diff = 1
    #this loop will keep applying the remove_adjacents() fn as long as the length of the list 
    #has gone down between one iteration and the next
    while diff > 0:
        #check length
        len_i = len(x)

        #apply fn
        x = remove_adjacents(x,opns)

        pairless = x
        
        #now check length again:
        diff = len_i - len(x)
    
    #now we have removed all adjacent pairs, loop over list and pick out the remaining closed brackets:
    corruptions = []
    for item in x:    
        if item in corr.values():
            corruptions.append(item)
    return corruptions, pairless

#Function to flip the order of a list, and the dirn of each char in list, giving the "completion" list (Part 2)
def complete(x):
    new = []
    for item in x:
        new.append(corr[item])
    new.reverse()
    return new

#Function to calculate the score of a given "completion" list (Part 2)
def scoring(x):
    scr = 0
    for item in x:
        scr *= 5
        scr += scores2[item]
    return scr

#Final Answer to Part 1:
#Now try the above function to each line of the example input and calculate the score
first_total = 0
for line in real_lines:
    if len(corrupted(line, opens)[0]) > 0:
        first_total += scores1[corrupted(line, opens)[0][0]]
print("Answer for Part 1: " + str(first_total))

#Final answer for Part 2:
#Calculate the score for each line of the puzzle input, and find the median value
pt2_scores = []
for line in real_lines:
    #only take the score into account if the corrupted fn returns an empty list of corruptions
    if len(corrupted(line,opens)[0]) == 0:
        pt2_scores.append(scoring(complete(corrupted(line,opens)[1])))
#Use statistics package to find median value:
print("Answer for Part 2: " + str(st.median(pt2_scores)))
```
